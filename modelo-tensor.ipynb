```python
import os
from PIL import Image
import numpy as np

base_folder_path = 'low_data/train'  # Cambia esto a la ruta de tu carpeta base

class_labels = {
    "neutral": 0,
    "frustration": 1
    # Agrega más clases y etiquetas según sea necesario
}

all_sequences = []
all_labels = []

for class_folder, label in class_labels.items():
    class_path = os.path.join(base_folder_path, class_folder)

    sequence_folders = [f for f in os.listdir(class_path) if os.path.isdir(os.path.join(class_path, f))]

    for sequence_folder in sequence_folders:
        sequence_path = os.path.join(class_path, sequence_folder)

        file_names = os.listdir(sequence_path)
        sorted_files = sorted(file_names, key=lambda x: int(os.path.splitext(x)[0]))

        images = []

        for file_name in sorted_files:
            file_path = os.path.join(sequence_path, file_name)
            image = Image.open(file_path)
            image_array = np.array(image)
            images.append(image_array)

        # Convertir la lista de imágenes a un arreglo numpy y agregarla a las secuencias totales
        images = np.array(images)  # (n_samples, height, width, channels)
        all_sequences.append(images)
        all_labels.append(label)

all_sequences = np.array(all_sequences)  # (n_sequences, n_samples, height, width, channels)
all_labels = np.array(all_labels)  # (n_sequences,)

print(all_sequences.shape)  # Verificar la forma del arreglo
print(all_labels.shape)  # Verificar la forma de las etiquetas
```

```python
import matplotlib.pyplot as plt

def visualize_sequence(sequence):
    fig, axes = plt.subplots(1, len(sequence), figsize=(10, 5))
    for i, img in enumerate(sequence):
        axes[i].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axes[i].axis('off')
    plt.show()
visualize_sequence(all_sequences[120])
```
